```python
#Load data using pandas 

import pandas as pd

veg_data = pd.read_csv('sri_lankan_vegetable_prices.csv')

veg_data
```

```python
#Print a summary of the dataset.

print(veg_data.describe())
print(veg_data.head())
```

```python
#Handle missing values

veg_data = veg_data.dropna()
```

```python
# Check for missing values in the entire dataset
print("Missing values :")
print(veg_data.isnull().sum())

if veg_data.isnull().sum().sum() == 0:
    print("The dataset is completely cleaned. No missing values!!.")
else:
    print("The dataset still contains missing values!!")
```

```python
from sklearn.preprocessing import LabelEncoder

# Convert 'Price Dates' column to datetime with the correct format
veg_data['Price Dates'] = pd.to_datetime(veg_data['Price Dates'], format='%Y-%m-%d')

veg_data['Month'] = veg_data['Price Dates'].dt.month
veg_data['Day_of_Week'] = veg_data['Price Dates'].dt.dayofweek

label_encoder = LabelEncoder()
veg_data['Month'] = label_encoder.fit_transform(veg_data['Month'])
veg_data['Day_of_Week'] = label_encoder.fit_transform(veg_data['Day_of_Week'])

print(veg_data.head())
```

```python
print(veg_data.info())
```

```python
metrics = veg_data[['Beans', 'Pumpkin', 'Carrot', 'Cabbage', 'Eggplant', 'Tomato', 'Potato', 'Onion']].agg(['mean', 'std'])
print(metrics)
```

```python
import matplotlib.pyplot as plt

veg_columns = ['Beans', 'Pumpkin', 'Carrot', 'Cabbage', 'Eggplant', 'Tomato', 'Potato', 'Onion']

# Plot each vegetable in a separate graph
for veg in veg_columns:
    plt.figure(figsize=(12, 3))
    plt.plot(veg_data['Price Dates'], veg_data[veg], label=veg, color='lightblue', marker='o')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.title(f'{veg} Price Trend')
    plt.legend()
    plt.grid(True) 
    plt.show()
```

```python
#Feature Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
veg_data[['Beans','Onion','Potato']] = scaler.fit_transform(veg_data[['Beans','Onion','Potato']])
```

```python
import seaborn as sns

sns.histplot(veg_data['Beans'], bins=20, kde=True)
plt.title('Beans Price Distribution')
plt.show()

sns.lineplot(x='Month', y='Beans', data=veg_data)
plt.title('Monthly Beans Price Trends')
plt.show()
```

```python
#Data analysis and visualization

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(veg_data['Average Price'], bins=20, kde=True)
plt.title('Average Price Distribution')
plt.show()

sns.lineplot(x='Month', y='Average Price', data=veg_data)
plt.title('Monthly Average Price Trends')
plt.show()
```

```python
#Summery Statistics

print(veg_data[['Beans','Onion']].mean())
print(veg_data[['Beans','Onion']].median())
print(veg_data[['Beans','Onion']].std())
```

```python
#Visualaization for mean, median, and std

mean_values = veg_data[['Beans', 'Onion']].mean()
median_values = veg_data[['Beans', 'Onion']].median()
std_values = veg_data[['Beans', 'Onion']].std()

labels = ['Beans', 'Onion']
x = range(len(labels))

plt.figure(figsize=(8,6))

plt.bar(x, mean_values, width=0.2, label='Mean', align='center')
plt.bar(x, median_values, width=0.2, label='Median', align='edge')
plt.bar(x, std_values, width=0.2, label='Std Dev', align='edge')

plt.xlabel('Vegetables')
plt.ylabel('Values')
plt.title('Mean, Median, and Standard Deviation for Beans and Onion')
plt.xticks(x, labels)
plt.legend()

plt.show()
```

```python
#Model Selection

from sklearn.model_selection import train_test_split

x = veg_data[['Month','Day_of_Week']]
y = veg_data['Average Price']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42) # Train 80%, Test 20%
```

```python
#Train models

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

#LinearRegression
lr = LinearRegression().fit(x_train, y_train)

#Random Forest
rf = RandomForestRegressor().fit(x_train, y_train)

#svm
svm = SVR().fit(x_train, y_train)
```

```python
# Regression Visualization for Linear Regression

plt.figure(figsize=(10, 6))
sns.scatterplot(x=x_train['Month'], y=y_train, color='blue', label='Actual (Train)')
sns.lineplot(x=x_train['Month'], y=lr.predict(x_train), color='red', label='Regression Line (Linear Regression)')
plt.title('Regression Line for Average Price vs Month (Linear Regression)')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.legend()
plt.show()
```

```python
# Regression Visualization for Random Forest

plt.figure(figsize=(10, 6))
sns.scatterplot(x=x_train['Month'], y=y_train, color='blue', label='Actual (Train)')
sns.scatterplot(x=x_train['Month'], y=rf.predict(x_train), color='green', label='Predictions (Random Forest)')
plt.title('Regression Predictions for Average Price vs Month (Random Forest)')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.legend()
plt.show()
```

```python
#Visualize Predict

plt.plot(y_test.values, label='Actual')
plt.plot(rf.predict(x_test), label='Predicted')
plt.title('Actual vs Predicted')
plt.legend()
plt.show()
```

```python
#Model Evalution

from sklearn.metrics import mean_squared_error, r2_score

for model, name in zip([lr,rf,svm], ['Linear Regression','Random Forest','SVM']):
    y_pred = model.predict(x_test)
    print(f"{name} - RÂ² Score: {r2_score(y_test, y_pred)}, MSE: {mean_squared_error(y_test, y_pred)}")
```
